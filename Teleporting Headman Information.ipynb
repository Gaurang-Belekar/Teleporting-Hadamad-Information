```python
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3) #Quantum Circuit with 3 qubit and 3 classical bit
```

```python
%matplotlib inline
```

```python
circuit.h(0)
```

```python
#Input
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
```

```python
# Creating EPR pair of medium and destination (i.e entagling q1 and q2)

#Creating EPR pair through bell circuit

circuit.h(1)
circuit.cx(1,2)
```

```python
#We have entangled q1 and q2. Now with the help of q1,we will teleport q0 information to q2.
#So first we will measure the entagled q1 state with information of q0

# For this we will use bell circuit measurement

circuit.cx(0,1)
circuit.h(0)
```

```python
circuit.draw(output='mpl')
```

```python
circuit.barrier()
```

```python
#Now after doing bell measurement, we will now measure the info of q0 and q1 ,and store it in classical bits
circuit.measure([0,1], [0,1])
```

```python
circuit.draw(output = 'mpl')
```

```python
#Now we will pass the information of q0 to q2 by measuring entangled q1, with q0

circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
```

```python
circuit.barrier()
```

```python
circuit.measure(2,2)
```

```python
circuit.draw(output ='mpl')
```

```python
#Plotting down the measurement of q2
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit,backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
#Here the output or the measurement of q2, i.e the last bit measured
# The Probablity of q2 = |0> sums up to 0.499 for the first four measurement out of eight
# While the Probablity of q2 = |1> sums up to 0. for the first four measurement out of eight
```
